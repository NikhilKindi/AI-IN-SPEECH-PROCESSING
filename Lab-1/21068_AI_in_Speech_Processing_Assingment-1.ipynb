# A1.Load the recorded speech file into your python workspace. Once loaded, plot the graph for the speech signal.You may use the below code from librosa asa reference.>>> y, sr = librosa.load(filename)>>>librosa.display.waveshow(y)

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

y, sr = librosa.load('nikhil.wav')
plt.figure(figsize=(15, 5))
librosa.display.waveshow(y)
plt.title('Speech Signal')


plt.axvspan(xmin=0.72, xmax=1.15, color='brown', alpha=0.3, label='AI')
plt.axvspan(xmin=1.21, xmax=1.4, color='green', alpha=0.3, label='in')
plt.axvspan(xmin=1.42, xmax=1.8, color='orange', alpha=0.3, label='Speech')
plt.axvspan(xmin=1.9, xmax=2.4, color='yellow', alpha=0.3, label='Processing')

plt.legend()
plt.show()
```

# A2. Observe the length & magnitude range of the signal from the plot. Observing the plot, relate to the spoken words and silence between the words.

```python
print("Length of the signal:",len(y))
print("Magnitude range:",np.min(y),"to", np.max(y))
magnitude_range = librosa.db_to_amplitude(np.max(librosa.amplitude_to_db(np.abs(y))))
print(f"Magnitude Range in dB: {np.max(magnitude_range):.2f} dB")
```

# A3. Take a small segment of the signal and play it. 

```python
import IPython.display as ip

start = int(0.7 * sr)
end = int(2.75 * sr)
segment = y[start:end]
plt.figure(figsize=(15,5))
librosa.display.waveshow(segment,color='red')
plt.title("Segment")
plt.show()
ip.Audio(segment, rate=sr)
```
